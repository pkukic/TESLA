```python
from dolfin import plot
import numpy as np
import matplotlib
matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt

from trainer import Trainer, Unit
```

```python
run_1, run_2 = './run_1', './run_2'
```

```python
import os

def trainers_from_run(run_path):
    return [{'path': os.path.join(run_path, file), 'trainer': Trainer.load(os.path.join(run_path, file))} for file in os.listdir(run_path)]
```

```python
run_1_trainers = trainers_from_run(run_1)
run_2_trainers = trainers_from_run(run_2)
```

```python
for d in run_1_trainers:
    name = d['path']
    trainer = d['trainer']
    print(f'Trainer {name}')
    for u in trainer.population:
        print(f'{u} goodness={u.goodness}')
```

```python
for d in run_2_trainers:
    name = d['path']
    trainer = d['trainer']
    print(f'Trainer {name}')
    for u in trainer.population:
        print(f'{u} goodness={u.goodness}')
```

```python
run_1_trainers_with_best_units = run_1_trainers[:]
lim = 8
for i, d in enumerate(run_1_trainers):
    name = d['path']
    trainer = d['trainer']
    print(f'Trainer {name}')
    best_units = []
    for u in trainer.population:
        print(f'{u} goodness={u.goodness}, best={u.goodness > lim}')
        if u.goodness > lim:
            best_units.append(u)
    run_1_trainers_with_best_units[i]['best_units'] = best_units
```

```python
run_2_trainers_with_best_units = run_2_trainers[:]
lim = 8
for i, d in enumerate(run_2_trainers):
    name = d['path']
    trainer = d['trainer']
    print(f'Trainer {name}')
    best_units = []
    for u in trainer.population:
        print(f'{u} goodness={u.goodness}, best={u.goodness > lim}')
        if u.goodness > lim:
            best_units.append(u)
    run_2_trainers_with_best_units[i]['best_units'] = best_units
```

```python
for d in run_1_trainers_with_best_units:
    name = d['path']
    trainer = d['trainer']
    best_units = d['best_units']
    print(f'Trainer {name}')
    for u in best_units:
        print(f'{u} goodness={u.goodness}')
```

```python
for d in run_2_trainers_with_best_units:
    name = d['path']
    trainer = d['trainer']
    best_units = d['best_units']
    print(f'Trainer {name}')
    for u in best_units:
        print(f'{u} goodness={u.goodness}')
```

```python
run_1_best_height_tuples = [[d['best_units'][i].heights_tuple for i in range(len(d['best_units']))] for d in run_1_trainers_with_best_units]
run_1_best_height_tuples = sum(run_1_best_height_tuples, [])
run_1_best_height_tuples = list(set(run_1_best_height_tuples))
```

```python
run_1_best_height_tuples
```

```python
run_2_best_height_tuples = [[d['best_units'][i].heights_tuple for i in range(len(d['best_units']))] for d in run_2_trainers_with_best_units]
run_2_best_height_tuples = sum(run_2_best_height_tuples, [])
run_2_best_height_tuples = list(set(run_2_best_height_tuples))
```

```python
run_2_best_height_tuples
```

```python
import domain

for i in range(len(run_1_best_height_tuples)):
    height_tup = run_1_best_height_tuples[i]
    mesh = domain.mesh(height_tup)
    plt.figure().set_figwidth(16)
    plot(mesh)
    ax = plt.gca()
    ax.set_aspect(1.6)
    ax.set_xlim([0, 700])
    ax.set_ylim([0, 50])
    ax.set_yticks([0, 10, 20, 30, 40, 50])
    ax.set_xticks([0, 100, 150, 200, 250, 300, 400, 450, 500, 550, 600, 700])
    extra_lines = [150, 250, 450, 550]
    for line in extra_lines:
        plt.axvline(x=line, color='black')
    plt.xlabel(r"x [$\AA$]")
    plt.ylabel(r"y [$\AA$]")
    plt.savefig(f"run_1_plots/run_1_mesh_{''.join([str(h) for h in height_tup])}.png",bbox_inches='tight', dpi=1200)
    plt.clf()
    plt.close()
    # plt.show()
```
